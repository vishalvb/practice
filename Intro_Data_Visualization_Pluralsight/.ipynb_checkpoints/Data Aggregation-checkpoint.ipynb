```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('obama.csv',parse_dates=['year_month'])#pandas will interpret year_month column as date/timestamps
data.head()
```

```python
plt.plot(data.year_month, data.approve_percent, 'o',markersize = 2, alpha = 0.3)#markersize is to set size of dots, 
                                                                                #alpha is to set transparency of dots
plt.show()
```

```python
data.shape
```

```python
data_mean = data.groupby('year_month').mean()
data_median = data.groupby('year_month').median()
```

```python
plt.plot(data_mean.index, data_mean.approve_percent,'red')
plt.plot(data_median.index, data_median.approve_percent,'green')
plt.legend(['Mean','Median'])
plt.plot(data.year_month, data.approve_percent, 'o',markersize = 2, alpha = 0.3)
plt.show()
```

```python
data_25 = data.groupby('year_month').quantile(0.25)#this finds 25% of approval ratings for each month
data_75 = data.groupby('year_month').quantile(0.27)#this finds 25% of approval ratings for each month


plt.plot(data_75.index, data_75.approve_percent,'green')
plt.plot(data_median.index, data_median.approve_percent,'red')
plt.plot(data_25.index, data_mean.approve_percent,'blue')
#plt.plot(data_median.index, data_median.approve_percent,'green')
plt.legend(['75th','Median','25th'])
plt.plot(data.year_month, data.approve_percent, 'o',markersize = 2, alpha = 0.3)
plt.show()
```
