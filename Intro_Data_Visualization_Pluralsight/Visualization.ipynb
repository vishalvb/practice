## Working with Histograms

```python
set(countries.continent)
```

```python
data_2007 = countries[countries.year == 2007]
```

```python
data_asia = data_2007[data_2007.continent == 'Asia']
data_europe = data_2007[data_2007.continent == 'Europe']
```

```python
print(len(set(data_asia.country)))
print(len(set(data_europe.country)))
```

```python
print('MEAN GDP per capita in Asia',data_asia.gdpPerCapita.mean())
print()
print('MEAN GDP per capita in Europe',data_europe.gdpPerCapita.mean())
print()
print('MEDIAN GDP per capita in Asia',data_asia.gdpPerCapita.median())
print()
print('MEDIAN GDP per capita in Europe',data_europe.gdpPerCapita.median())
plt.print()
```

```python
plt.subplot(2,1,1)#2 rows and 1 column with 1 = select the first figure
#range is used to compare two plots with same x-axis
plt.hist(data_asia.gdpPerCapita,20, edgecolor = 'black',range=(0,50000))
plt.ylabel('Asia')
plt.xlabel('GDPperCaptia')
plt.subplot(2,1,2)#2 rows and 1 column with 2 = select the second figure
plt.ylabel('Europe')
plt.xlabel('GDPperCaptia')
plt.hist(data_europe.gdpPerCapita,20, edgecolor = 'black',range=(0,50000))
plt.show()
```
